```python
import pandas as pd
import numpy as np 
import os
import requests
```

```python
# Trazendo tabela do wikipedia 
requisicao = requests.get("https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_2017_-_S%C3%A9rie_A")
tabelas = pd.read_html(requisicao.text)
```

```python
# Definindo tabelas a partir de lista de tabelas
df_artilheiros = tabelas[11]
df_assistencias = tabelas[12]
```

```python
path= os.getcwd()
path_excel= os.path.join(path,'brasileirao2017.xlsx')
```

```python
with pd.ExcelWriter(path_excel) as writer:  
    df_artilheiros.to_excel(writer, sheet_name='artilheiros',index=False)
    df_assistencias.to_excel(writer, sheet_name='assistencias',index=False)
```

```python
df_artilheiros["posicao"] = df_artilheiros.index +1
df_artilheiros["posicao"] = df_artilheiros["posicao"].astype('str')+"°"

df_assistencias["posicao"] = df_assistencias.index +1
df_assistencias["posicao"] = df_assistencias["posicao"].astype('str')+"°"

df_artilheiros = df_artilheiros.rename(columns={'Jogador': 'jogador_gol','Time': 'time_gol'})
df_assistencias = df_assistencias.rename(columns={'Jogador': 'jogador_ass', 'Time':'time_ass'})
```
